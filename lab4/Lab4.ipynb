### Импорты

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

### Исходное изображение

```python
image = cv.imread('images/lab4_img.jpg', cv.IMREAD_GRAYSCALE)
plt.figure(figsize=(14, 10))
plt.imshow(image, cmap='gray')
```

### Бинаризация

```python
_, image = cv.threshold(image, 120, 255, cv.THRESH_BINARY)
    
plt.figure(figsize=(14, 10))
plt.imshow(image, cmap='gray')
```
